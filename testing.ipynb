```python
import tensorflow as tf
import os
import numpy as np
import matplotlib.pyplot as plt
```

# A MEAN experiment

This notebook is a little experiment to get a neural network to learn how to find the mean of a set of data. The data is question is drawn from a normal distribution.

```python
num_data = 1_000_000
samples_per_datum = 100
mean_range = 200

means = np.random.randint(-mean_range, mean_range+1, size=num_data)
data = np.array([np.random.normal(mu, 1, size=samples_per_datum) for mu in means])

selector = np.array([False]*num_data)
indices = np.random.choice(num_data, int(.85*num_data), replace=False)
selector[indices] = True

train_input, train_target = data[selector], means[selector]

deselector = np.logical_not(selector)
test_input, test_target = data[deselector], means[deselector]

```

```python
# The model is a fairly standard Sequential one, the layers were not chosen for a particular reason other than for correct input and output dimensions.

model = tf.keras.models.Sequential([
  tf.keras.Input(shape=samples_per_datum),
  tf.keras.layers.Dense(128, activation='elu'),
  tf.keras.layers.Dense(64, activation='sigmoid'),
  tf.keras.layers.Dense(1, activation='linear')
])
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

model.fit(train_input, train_target, epochs=100, batch_size=1000)

print(model.evaluate(test_input, test_target, verbose=2))
```

```python
a = np.random.normal(12.3456789, 1, size=(1,samples_per_datum))

a_mean = a.mean()
print(a_mean)

model_mean = model(a)

print(tf.keras.losses.mae(a_mean, model_mean))
```
